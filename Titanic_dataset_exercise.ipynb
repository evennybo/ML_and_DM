# Question #13 Create categories by age as follows:

1. Between 0 and 18: Children
2. Between 19 and 25: Youth
3. Between 26 and 35: YoungAdult
4. Between 36 and 60: MiddleAged
5. Over 60: Senior

Then, find how many passengers belong to the Youth category

```python
bins = [0,18, 25, 35, 60, 80]
group_names = ['Children','Youth', 'YoungAdult', 'MiddleAged', 'Senior']
age_cats= pd.cut(df['Age'], bins=bins, labels=group_names)
age_cats
```

```python
age_cats.value_counts()
```

# Question #14 Find the survival rate by passenger's gender, then sort the result by survived in descending order.

>Condition: use the mean() function

```python
df[["Gender", "Survived"]].groupby(['Gender'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```

# Question #15 Make two data frames first. The first data frame consists of PassengerId and Name columns. The second data frame consists of Name, SibSp, Parch, and Embarked. Then, merge these two data frames consisting of PassengerId, Name, SibSp, Parch, and Embarked. Find how many passengers were on board alone and with family members.

```python
df1 = df.drop(columns=['Survived', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Ticket', 'Fare', 'Cabin', 'Embarked'])
df1
df2 = df.drop(columns=['PassengerId', 'Survived', 'Pclass', 'Sex', 'Age', 'Ticket', 'Fare', 'Cabin'])
df2
```

```python
df3 = pd.merge(df1, df2)
df3
```

```python
df3['FamilySize'] = df3['SibSp'] + df3['Parch'] + 1
df3.loc[df3['FamilySize']==1, ['PassengerId', 'Name', 'FamilySize']]
```

```python
df3['FamilySize'] = df3['SibSp'] + df3['Parch'] + 1
df3.loc[df3['FamilySize']>1, ['PassengerId', 'Name', 'FamilySize']]
```
